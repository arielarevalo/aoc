# Advent of Code 2025, Day 1

```python
with open('input.txt', 'r') as f:
    puzzle_input = f.read().strip()

text_rotations = puzzle_input.splitlines()
scalar_rotations = [rotation.replace('L', '-').replace('R', '') for rotation in text_rotations]
rotations = [int(rotation) for rotation in scalar_rotations]
```

## [First Puzzle:](https://adventofcode.com/2025/day/1)

```python
dial = 50
zeroes_count = 0
for rotation in rotations:
    dial = (dial + rotation) % 100
    if dial == 0:
        zeroes_count += 1

zeroes_count
```

```python
from itertools import accumulate
dials = accumulate([50] + rotations, lambda x, y: (x + y) % 100)
sum(1 for value in dials if value == 0)
```

## [Second Puzzle:](https://adventofcode.com/2025/day/1/#part2)

```python
dial = 50
zeroes_count = 0
for rotation in rotations:
    start = dial if rotation > 0 else -dial % 100
    zeroes_count += (start + abs(rotation)) // 100
    dial = (dial + rotation) % 100

zeroes_count
```

```python
from itertools import accumulate
dials = accumulate([50] + rotations, lambda x, y: (x + y) % 100)
sum(((dial if rotation > 0 else -dial % 100) + abs(rotation)) // 100 for dial, rotation in zip(dials, rotations))
```
