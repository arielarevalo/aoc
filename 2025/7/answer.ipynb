# Advent of Code 2025, Day 7

```python
with open('input.txt', 'r') as f:
    puzzle_input = f.read().strip()

diagram = [list(line) for line in puzzle_input.split('\n')]
```

```python
SOURCE = 'S'
SPLITTER = '^'
BEAM = '|'
```

## [First Puzzle:](https://adventofcode.com/2025/day/7)

```python
split_count = 0
for row_idx in range(len(diagram) - 1):
    next_row_idx = row_idx + 1

    row = diagram[row_idx]
    next_row = diagram[next_row_idx]

    if SOURCE in row:
        col = row.index(SOURCE)
        next_row[col] = BEAM
        continue

    beam_indices = [i for i, v in enumerate(row) if v == BEAM]

    for col in beam_indices:
        if next_row[col] == SPLITTER:
            next_row[col - 1] = BEAM
            next_row[col + 1] = BEAM
            split_count += 1
        else:
            next_row[col] = BEAM

split_count
```

## [Second Puzzle:](https://adventofcode.com/2025/day/7/#part2)

```python
# I could just dump it into networkx and count paths to each leaf, but what's the fun in that?
```

```python
diagram = [list(line) for line in puzzle_input.split('\n')]

# Replace all '.' with '0'
for row_idx in range(len(diagram)):
    for col_idx in range(len(diagram[row_idx])):
        if diagram[row_idx][col_idx] == '.':
            diagram[row_idx][col_idx] = '0'
```

```python
EMPTY = '0'
SYMBOLS = [SOURCE, EMPTY, SPLITTER]
```

```python
for row_idx in range(len(diagram) - 1):
    next_row_idx = row_idx + 1

    row = diagram[row_idx]
    next_row = diagram[next_row_idx]

    if SOURCE in row:
        col = row.index(SOURCE)
        next_row[col] = '1'
        continue

    beam_indices = [i for i, v in enumerate(row) if v not in SYMBOLS]

    for col in beam_indices:
        if next_row[col] == SPLITTER:
            next_row[col - 1] = str(int(next_row[col - 1]) + int(row[col]))
            next_row[col + 1] = str(int(next_row[col + 1]) + int(row[col]))
        else:
            next_row[col] = str(int(next_row[col]) + int(row[col]))

sum(int(v) for v in diagram[-1])
```
