# Advent of Code 2025, Day 5

```python
with open('input.txt', 'r') as f:
    puzzle_input = f.read().strip()

ingredient_ids = puzzle_input.split('\n')
blank_separator_idx = ingredient_ids.index('')
fresh_ingredient_id_ranges = [(int(text_range.split('-')[0]), int(text_range.split('-')[1])) for text_range in ingredient_ids[:blank_separator_idx]]
available_ingredient_ids = list(map(int, ingredient_ids[blank_separator_idx + 1:]))
```

## [First Puzzle:](https://adventofcode.com/2025/day/5)

```python
fresh_ingredient_ids = set()
for ingredient_id in available_ingredient_ids:
    for start_range, end_range in fresh_ingredient_id_ranges:
        if start_range <= ingredient_id <= end_range:
            fresh_ingredient_ids.add(ingredient_id)
            break
```

```python
len(fresh_ingredient_ids)
```

## [Second Puzzle:](https://adventofcode.com/2025/day/5/#part2)

```python
fresh_ingredient_id_ranges.sort(key=lambda r: r[0])

final_ranges = list()
final_ranges.append(fresh_ingredient_id_ranges[0])

for current_start, current_end in fresh_ingredient_id_ranges[1:]:
    last_end = final_ranges[-1][1]

    if current_start <= last_end:
        final_ranges[-1] = (final_ranges[-1][0], max(last_end, current_end))
    else:
        final_ranges.append((current_start, current_end))
```

```python
sum([range[1] - range[0] + 1 for range in final_ranges])
```
