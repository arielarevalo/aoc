# Advent of Code 2025, Day 6

```python
with open('input.txt', 'r') as f:
    puzzle_input = f.read().strip()
```

## [First Puzzle:](https://adventofcode.com/2025/day/6)

```python
import pandas as pd
import io

homework = pd.read_csv(io.StringIO(puzzle_input), sep='\\s+', header=None, engine='python')
homework_t = homework.transpose()

# Cast all columns except the last to int
for col in homework_t.columns[:-1]:
    homework_t[col] = homework_t[col].astype(int)

# Add a column to the end that conditionally (based on whether the last column is '+' or '*') adds or multiplies all the values in the previous columns
homework_t['result'] = homework_t.apply(lambda row: row[:-1].sum() if row.iloc[-1] == '+' else row[:-1].prod(), axis=1)
homework_t['result'].sum()
```

## [Second Puzzle:](https://adventofcode.com/2025/day/6/#part2)

```python
import operator
import functools

homework = pd.DataFrame([list(line) for line in puzzle_input.split('\n')])
homework_t = homework.transpose()

homework_t['values'] = homework_t.apply(lambda row: ''.join(row[:-1].astype(str)), axis=1).str.strip()
operations = homework_t.rename(columns={4: 'operator'}).drop(columns=[0, 1, 2, 3])

operators = operations['operator'].replace(' ', pd.NA).dropna().tolist()
operators = [operator.add if op == '+' else operator.mul for op in operators]

values = operations['values'].replace('', 'SEP').tolist()
value_lists = [[int(num) for num in chunk.strip().split(" ")] for chunk in " ".join(values).split("SEP")]

operations = zip(value_lists, operators)
results = [functools.reduce(op, nums) for nums, op in operations]
```

```python
sum(results)
```
