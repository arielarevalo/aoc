# Advent of Code 2025, Day 4

```python
with open('input.txt', 'r') as f:
    puzzle_input = f.read().strip()

grid = [list(row) for row in puzzle_input.split('\n')]
```

## [First Puzzle:](https://adventofcode.com/2025/day/4)

```python
# 0.025s

import time
start_time = time.time()

roll_locations = set([(r, c) for r, row in enumerate(grid) for c, cell in enumerate(row) if cell == '@'])
roll_location_num_neighbors = {loc: sum(
    1 for dr in (-1, 0, 1) for dc in (-1, 0, 1)
    if (dr != 0 or dc != 0) 
        and 0 <= loc[0] + dr < len(grid)
        and 0 <= loc[1] + dc < len(grid[0])
        and grid[loc[0] + dr][loc[1] + dc] == '@'
) for loc in roll_locations}

end_time = time.time()
print(f"Execution time: {end_time - start_time} seconds")
```

```python
len(list(loc for loc, num_neighbors in roll_location_num_neighbors.items() if num_neighbors < 4))
```

## [Second Puzzle:](https://adventofcode.com/2025/day/4/#part2)

```python
total_removed = 0
while True:
    print(f"Current roll locations: {len(roll_locations)}")
    removable_roll_locations = set([loc for loc, num_neighbors in roll_location_num_neighbors.items() if num_neighbors < 4])
    if len(removable_roll_locations) == 0:
        break
    total_removed += len(removable_roll_locations)
    print(f"Removing {len(removable_roll_locations)} roll locations")
    roll_locations -= removable_roll_locations
    for loc in removable_roll_locations:
        grid[loc[0]][loc[1]] = '.'
    roll_location_num_neighbors = {loc: sum(
        1 for dr in (-1, 0, 1) for dc in (-1, 0, 1)
        if (dr != 0 or dc != 0) 
            and 0 <= loc[0] + dr < len(grid)
            and 0 <= loc[1] + dc < len(grid[0])
            and grid[loc[0] + dr][loc[1] + dc] == '@'
    ) for loc in roll_locations}
```
